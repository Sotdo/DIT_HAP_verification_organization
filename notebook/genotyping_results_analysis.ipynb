```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../results/batch_genotyping_results.xlsx")
```

```python
def calculate_deletion_fraction(group):
    deletion = group.query("genotype == 'Deletion'").dropna(subset=['genotype'])
    deletion_fraction = len(deletion) / len(group.dropna(subset=['genotype']))
    return deletion_fraction
filtered_data = data.groupby(
    ["round", "gene_num", "gene_name", "colony_id", "date"]
).filter(
    lambda x: 0.4 <= calculate_deletion_fraction(x) <= 0.6
)
```

```python
def calculate_deletion_area_fraction(group):
    deletion = group.query("genotype == 'Deletion'").dropna(subset=['genotype'])
    wt = group.query("genotype == 'WT'").dropna(subset=['genotype'])
    deletion_area = deletion.filter(like='area_').sort_index(axis=1)
    wt_area = wt.filter(like='area_').sort_index(axis=1)
    res = deletion_area.median() / wt_area.median()
    return res

area_fraction = filtered_data.groupby(
    ["round", "gene_num", "gene_name", "colony_id", "date"]
).apply(
    calculate_deletion_area_fraction,
    include_groups = True
).dropna().groupby(
    ["round", "gene_num", "gene_name"]
).filter(
    lambda x: len(x) == 2
)
```
